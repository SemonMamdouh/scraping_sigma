```python
import requests
from bs4 import BeautifulSoup

with open('sigma1.csv' , mode = 'w', encoding= 'utf_8') as fd :
    #header
    fd.write('title_product,price,model,brand,seller,viewed\n')
    #request to get data 

    res= requests.get('https://www.sigma-computer.com/search?search=laptop&submit_search=&route=product%2Fsearch')
    soup = BeautifulSoup(res.text , 'html.parser')

    #find all product
    all_divs =soup.find_all('div' , attrs={'class':'product-layout col-lg-15 col-md-4 col-sm-6 col-xs-12'})

    for i in range(len(all_divs)) :
        print(f'scraping element {i+1}')
        
        #avaliable or not
        avaliability = all_divs[i].find('span' , attrs = {'class':['stock','stock_N']}).text.strip(' ')

        #link_of_each_product
        
        laptop_link = 'https://www.sigma-computer.com/' + all_divs[i].find('div', attrs = {'class':'caption hide-cont'}).find('a').attrs['href']
        
        while True :
            re =requests.get(laptop_link)
            if re  :
                break


        soup1 =BeautifulSoup(re.text , 'html.parser')
        content_div = soup1.find('div' , attrs = {'class':'content-product-right col-md-7 col-sm-12 col-xs-12'})

        #title_product
        title_product = content_div.find('div' , attrs = {'class' :'title-product'}).text.replace('\n','')

        #price
        price = content_div.find('div' , attrs = {'class':'product_page_price price'}).text.replace('\n', '')
        
        #model
        model = content_div.find('div' , attrs = {'class':'model'}).text.split(":")[1].strip()

        #brand
        model_details = content_div.find_all('h3' ,attrs ={'class' :'modtitle'})
        brand = model_details[0].find('a').text.strip()

        #seller
        seller = model_details[2].find('a').text

        #views
        viewed= model_details[3].text.split(':')[1].strip()

        fd.write(f'\"{title_product}\",\"{price}\",\"{model}\",\"{brand}\",\"{seller}\",\"{viewed}\"\n')
        
                
```

```python
import requests
from bs4 import BeautifulSoup

with open('sigma2.csv' , mode = 'w', encoding= 'utf_8') as fd :
    #header
    fd.write('title_product , price , model , brand ,seller , viewed ,info_1 ,info_2, info_3 \n')
    #request to get data 

    res= requests.get('https://www.sigma-computer.com/search?search=laptop&submit_search=&route=product%2Fsearch')
    soup = BeautifulSoup(res.text , 'html.parser')

    #find all product
    all_divs =soup.find_all('div' , attrs={'class':'product-layout col-lg-15 col-md-4 col-sm-6 col-xs-12'})

    for i in range(50) :
        print(f'scraping element {i+1}')
        
        #avaliable or not
        avaliability = all_divs[i].find('span' , attrs = {'class':['stock','stock_N']}).text.strip(' ')

        #link_of_each_product
        laptop_link = 'https://www.sigma-computer.com/' + all_divs[i].find('div', attrs = {'class':'caption hide-cont'}).find('a').attrs['href']
        
        while True :
            re =requests.get(laptop_link)
            if re  :
                break
            
        soup1 =BeautifulSoup(re.text , 'html.parser')
        content_div = soup1.find('div' , attrs = {'class':'content-product-right col-md-7 col-sm-12 col-xs-12'})

        #title_product
        title_product = content_div.find('div' , attrs = {'class' :'title-product'}).text.replace('\n','')

        #price
        price = content_div.find('div' , attrs = {'class':'product_page_price price'}).text.replace('\n', '')
        
        #model
        model = content_div.find('div' , attrs = {'class':'model'}).text.split(":")[1].strip()

        #brand
        model_details = content_div.find_all('h3' ,attrs ={'class' :'modtitle'})
        brand = model_details[0].find('a').text.strip()

        #seller
        seller = model_details[2].find('a').text

        #views
        viewed= model_details[3].text.split(':')[1].strip()

        # info 
        info_elements = []
        for j in range(3) :
            if model_details[-1].attrs == {'class': ['modtitle']}:
                info = model_details[-1].find_all(['li', 'span', 'div','p'])
                for e in info :
                    if e.text in ['KeyFeatures:','Key Features:','Battery:','Processor:' ,'Graphics Card:','Memory:','Storage:','Display:','OS:','Build:','Resolution & Performance:','Build Quality:','Advanced Features:','Compatibility:','Cable Length:' ,'Cores/Threads:','Cache:','RAM:'] :
                        info.remove(e)
                if j < len(info):
                    text = info[j].text.strip('\n').replace('\xa0 \xa0' ,'').replace('\r\n','').replace(',' ,' ').replace('\n' , '').replace('"' , "'")
                    if text:  
                        info_elements.append(text.strip('\xa0').strip('\n'))
                    else :
                        info_elements.append('none')
                else :
                    info_elements.append('none')
            else :
                info_elements.append('none')

        # check if elements in list not space or empty
        for i in range(len(info_elements)) :
            if  info_elements[i] == '' or info_elements[i]  == ' ':
                info_elements[i] =='none'
        

        fd.write(f"\"{title_product}\",\"{price}\",\"{model}\",\"{brand}\",\"{seller}\",\"{viewed}\",\"{info_elements[0]}\",\"{info_elements[1]}\",\"{info_elements[2]}\"\n")
```
